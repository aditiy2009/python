```
import numpy as np
import matplotlib.pyplot as plt

plt.quiver(0,0,3,4)
plt.show()
```

```
#the vector scale is not perfect
plt.quiver(0,0,3,4, scale_units='xy' ,angles='xy' ,scale=1)
plt.show()
```

```
#we can see that the vector is going out  of range of plot to adjust that we need to change the limits of axis

plt.quiver(0,0,3,4, scale_units='xy',angles='xy',scale=1 )
plt.xlim(0,10)
plt.ylim(0,10)
plt.show()
```

```
#plotting multiple vectors

plt.quiver(0,0,3,4,scale_units='xy',angles='xy',scale=1)
plt.quiver(0,0,-3,4,scale_units='xy',angles='xy',scale=1)
plt.quiver(0,0,3,-4,scale_units='xy',angles='xy',scale=1)
plt.quiver(0,0,-3,-4,scale_units='xy',angles='xy',scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```
#if you want vector with different colors 
#plotting multiple vectors

plt.quiver(0,0,3,4,scale_units='xy',angles='xy',scale=1,color='r')
plt.quiver(0,0,-3,4,scale_units='xy',angles='xy',scale=1,color='g')
plt.quiver(0,0,3,-4,scale_units='xy',angles='xy',scale=1,color='b')
plt.quiver(0,0,-3,-4,scale_units='xy',angles='xy',scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

#Function to print vectors in one-go

```
import numpy as np
import matplotlib.pyplot as plt


def vector_one_go(v):
  
  colors=['r','g','y','b']
  c=0
  for i in v:
    
    plt.quiver(i[0],i[1],i[2],i[3],scale_units='xy',angles='xy',scale=1,color=colors[c%len(colors)])
   
    c=c+1
    
  plt.xlim(-10,10)
  plt.ylim(-10,10)
  plt.show()
  

  
```

```
vector_one_go([[0,0,3,-4],[0,0,3,4]]) #passing parameter to function
```

```
vec=[np.asarray([0,0,3,-4]),np.asarray([0,0,3,4])] #created list of numpy array as to perform computations
```

```
vector_one_go([vec[0],vec[1]])
```

```
vec[0]+vec[1] 
```

```
vector_one_go([vec[0],vec[1],vec[0]+vec[1]]) #resultant vector is long when angle is less
```

```
vector_one_go([vec[0],vec[1],vec[0]-vec[1]]) # take mirror image of vec[0] and then calculate the answer
```

#Dot product

```
a=np.asarray([3,4])
b=np.asarray([4,-1])
```

```
a_b_dot=np.dot(a,b)
```

```
print(a_b_dot)
```

#Projection of vectors a on b

```
a_b_project=a_b_dot/np.linalg.norm(b)
```

```
a_b_project
```

```
dirct=(a_b_project/np.linalg.norm(b))*b
```

```
dirct
```
